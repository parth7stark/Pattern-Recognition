```python

'''
Q2. Calculate the distance between the two normalized histograms H1 and H2 using each of the following methods:
    I. KL Distance
    II. Bhattacharyya Distance
    H1 = [ 0.24, 0.2, 0.16, 0.12, 0.08, 0.04, 0.12, 0.04]
    H2 = [ 0.22, 0.19, 0.16, 0.13, 0.11, 0.08, 0.05, 0.02]

'''
import numpy as np
import math

H1 = np.array([0.24, 0.2, 0.16, 0.12, 0.08, 0.04, 0.12, 0.04])
H2 = np.array([0.22, 0.19, 0.16, 0.13, 0.11, 0.08, 0.05, 0.06])



#calculating KL distance wrt H1
KL = H1/H2
KL = np.log(KL)
KL = KL*H1
KL_dist = np.sum(KL)

# print(KL)
print("\n KL distance(w.r.t H1): {}".format(KL_dist))


#calculating KL distance wrt H2

KL = H2/H1
KL = np.log(KL)
KL = KL*H2
KL_dist = np.sum(KL)

# print(KL)
print("\n KL distance(w.r.t H2): {}".format(KL_dist))



#calculating Bhattacharyya Distance
x = H1*H2
x = np.sqrt(x)
Bhatt_coeff = np.sum(x)

Bhatt_dist = -np.log(Bhatt_coeff)

print("\n Bhattacharyya distance: {}".format(Bhatt_dist))
```

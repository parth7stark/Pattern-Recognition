```python
'''
final Q1
i) Assume P(Ï‰1) =0.8; P(Ï‰2)=0.2 
ii)  Assume P(Ï‰1) = P(Ï‰2) 
 
Ï‰1 = [1,6; 3,4; 3,8; 5,6] Ï‰2 = [3,0; 1,-2;3,-4;5,-2] 

'''

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sympy import *
from sympy import plot, symbols
from sympy.plotting.plot import List2DSeries
%matplotlib notebook

x1 = [1,3,3,5]
y1 = [6,4,8,6] 

x2 = [3,1,3,5]
y2 = [0,-2,-4,-2]

w1 = pd.DataFrame({"x":x1, "y": y1 })
print(w1)

w2 = pd.DataFrame({"x":x2, "y":y2 })
print(w2)

P_w1 = 0.8
P_w2 = 0.2

#calculate covariance matrix for class w1
w1_x_mean = np.mean(w1["x"])
w1_y_mean = np.mean(w1["y"])

# w1_y_mean

w1_mean = np.array([[w1_x_mean,w1_y_mean]])
print(w1_mean)

w1_mean1 = np.tile(w1_mean,(len(w1),1))

#find Z matrix
w1_Z = w1 - w1_mean1
w1_Z

# #Covariance matrix
w1_cov = (np.dot(w1_Z.T, w1_Z))/(w1.shape[0]-1)
print("w1 Covariance: \n",w1_cov)

#Cov matrix inverse
w1_cov_inv = np.linalg.inv(w1_cov)
# w1_cov_inv

#calculate covariance matrix for class w2
w2_x_mean = np.mean(w2["x"])
w2_y_mean = np.mean(w2["y"])

# w2_y_mean

w2_mean = np.array([[w2_x_mean,w2_y_mean]])
print(w2_mean)

w2_mean1 = np.tile(w2_mean,(len(w2),1))

#find Z matrix
w2_Z = w2 - w2_mean1
w2_Z

# #Covariance matrix
w2_cov = (np.dot(w2_Z.T, w2_Z))/(w2.shape[0]-1)
print("w2 Covariance: \n",w2_cov)

#Cov matrix inverse
w2_cov_inv = np.linalg.inv(w2_cov)
# w1_cov_inv


#finding decision boundary

#checking case 1: 

#sigma^2 or variance
var_w1 = w1_cov[0][0]
if np.array_equal(w1_cov, (var_w1*np.eye(w1_cov.shape[0]))) and np.array_equal(w2_cov, (var_w1*np.eye(w2_cov.shape[0]))):
    print("case1")
elif np.array_equal(w1_cov, w2_cov): 
    print("case2")
else:
    print("Case3")
    
    
x,y = symbols('x y')
X=Matrix([[x,y]]).T


g1 = X.T.dot((-0.5*w1_cov_inv).dot(X)) + (np.dot(w1_cov_inv,w1_mean.T)).T.dot(X) -0.5*np.dot(w1_mean,np.dot(w1_cov_inv, w1_mean.T)) - 0.5*np.log(np.linalg.det(w1_cov))
g2 = X.T.dot((-0.5*w2_cov_inv).dot(X)) + (np.dot(w2_cov_inv,w2_mean.T)).T.dot(X) -0.5*np.dot(w2_mean,np.dot(w2_cov_inv, w2_mean.T)) - 0.5*np.log(np.linalg.det(w2_cov))

boundary = g1-g2
b = boundary[0][0]
print(b)


#plot boundary
plot_implicit(b,(x,-10,10), (y,-10,10))

plt.scatter(x1,y1, label = "w1")
plt.scatter(x2,y2, label = "w2")

# plt.legend()
# sympy.plot(x1,y1)
plt.show()
```
